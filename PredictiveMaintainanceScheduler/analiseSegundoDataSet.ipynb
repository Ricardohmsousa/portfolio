```python
import pandas as pandas
import matplotlib.pyplot as matplot
import seaborn as sns

trainDataframe = pandas.read_csv("./ALLtrainMescla5D.csv")
testDataframe = pandas.read_csv("./ALLtestMescla5D.csv")

trainDataframe = trainDataframe.drop(columns=['datetime','RUL'])
testDataframe = testDataframe.drop(columns=['datetime','RUL'])

print(trainDataframe)
```

```python
print("Max Cycle Histogram")

matplot.hist(trainDataframe.groupby("machineID",sort=False)["RUL_I"].max(), bins='sturges')
matplot.title('Max RUL Histogram')
matplot.xlabel('Maximum RUL')
matplot.ylabel('Frequency')
```

```python
fig, axs = matplot.subplots(6,6,figsize=(25, 25))
fig.suptitle("Sensor Data Histogram")
fig.tight_layout(pad=3.5)

columnSelect = [18,19,20,21,22,23,31,32]
dataSetRow = 3
for column in range(0,6):
    for row in range(0,6):
        if dataSetRow in columnSelect:
            selectedColumn = trainDataframe.iloc[:, dataSetRow]
            axs[column,row].hist(selectedColumn, bins='sturges')
            axs[column,row].set_title(trainDataframe.columns[dataSetRow])
            axs[column,row].set_yscale('log')
        elif dataSetRow <= 32:
            axs[column,row].hist(trainDataframe.iloc[:, dataSetRow], bins='sturges')
            axs[column,row].set_title(trainDataframe.columns[dataSetRow])
        else:
            axs[column,row].axis('off')
            
       
        
        dataSetRow +=1

matplot.show()
```

```python
correlationDataSensors = trainDataframe.iloc[:,2:]

correlationMatrixSensors = correlationDataSensors.corr()
fig, ax = matplot.subplots(figsize=(15, 15))
sns.heatmap(correlationMatrixSensors, annot = True,ax=ax,cmap="PiYG",fmt='.1f')

matplot.show()
```
